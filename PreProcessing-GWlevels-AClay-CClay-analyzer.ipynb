```python
# Merge df_perf and merged_df on 'site_code'
merged_df = pd.merge(df_perf, merged_df, on='site_code', how='inner')
merged_df.head()
```

```python
# Convert existing columns to numeric, forcing errors to NaN if conversion fails
numeric_columns = ['wlm_gse', 'top_prf_int', 'bot_prf_int', 'Depth_2top_Eclay', 'Thickness_Eclay',
                   'thickness_Aclay', 'Aclay_base_altitude', 'thickness_Cclay', 'Cclay_base_altitude',
                  'MidTulareClay_Depth2top','MidTulareClay_Thickness','AmnicolaClaystone-Depth2top','AmnicolaClaystone-thickness'
                  ]
merged_df[numeric_columns] = merged_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Perform calculations
merged_df['altitude_top_prf'] = merged_df['wlm_gse'] - merged_df['top_prf_int']
merged_df['altitude_bot_prf'] = merged_df['wlm_gse'] - merged_df['bot_prf_int']

merged_df['altitude_top_EClay'] = merged_df['wlm_gse'] - merged_df['Depth_2top_Eclay']
merged_df['altitude_base_EClay'] = (merged_df['wlm_gse'] - merged_df['Depth_2top_Eclay'] 
                                    - merged_df['Thickness_Eclay'])

merged_df['altitude_top_Aclay'] =  merged_df['Aclay_base_altitude']+ merged_df['thickness_Aclay']

merged_df['altitude_top_Cclay'] =  merged_df['Cclay_base_altitude'] + merged_df['thickness_Cclay']

merged_df['altitude_top_MidTulareClay'] = merged_df['wlm_gse'] - merged_df['MidTulareClay_Depth2top']
merged_df['altitude_base_MidTulareClay'] = (merged_df['wlm_gse'] - merged_df['MidTulareClay_Depth2top'] 
                                    - merged_df['MidTulareClay_Thickness'])

merged_df['altitude_top_AmnicolaClaystone'] = merged_df['wlm_gse'] - merged_df['AmnicolaClaystone-Depth2top']
merged_df['altitude_base_AmnicolaClaystone'] = (merged_df['wlm_gse'] - merged_df['AmnicolaClaystone-Depth2top'] 
                                    - merged_df['AmnicolaClaystone-thickness'])

print("Calculations completed successfully.")
```

```python
# Convert columns to numeric, forcing errors to NaN (in case there are any non-numeric values)
merged_df['top_prf_int'] = pd.to_numeric(merged_df['top_prf_int'], errors='coerce')
merged_df['Depth_2top_Eclay'] = pd.to_numeric(merged_df['Depth_2top_Eclay'], errors='coerce')
merged_df['Thickness_Eclay'] = pd.to_numeric(merged_df['Thickness_Eclay'], errors='coerce')

merged_df['altitude_top_EClay'] = pd.to_numeric(merged_df['altitude_top_EClay'], errors='coerce')
merged_df['altitude_base_EClay'] = pd.to_numeric(merged_df['altitude_base_EClay'], errors='coerce')

merged_df['altitude_top_Aclay'] = pd.to_numeric(merged_df['altitude_top_Aclay'], errors='coerce')
merged_df['Aclay_base_altitude'] = pd.to_numeric(merged_df['Aclay_base_altitude'], errors='coerce')

merged_df['altitude_top_Cclay'] = pd.to_numeric(merged_df['altitude_top_Cclay'], errors='coerce')
merged_df['Cclay_base_altitude'] = pd.to_numeric(merged_df['Cclay_base_altitude'], errors='coerce')

merged_df['altitude_top_MidTulareClay'] = pd.to_numeric(merged_df['altitude_top_MidTulareClay'], errors='coerce')
merged_df['altitude_base_MidTulareClay'] = pd.to_numeric(merged_df['altitude_base_MidTulareClay'], errors='coerce')

merged_df['altitude_top_AmnicolaClaystone'] = pd.to_numeric(merged_df['altitude_top_AmnicolaClaystone'], errors='coerce')
merged_df['altitude_base_AmnicolaClaystone'] = pd.to_numeric(merged_df['altitude_base_AmnicolaClaystone'], errors='coerce')

# Apply the condition to set 'Eclay' column
merged_df['Eclay'] = merged_df.apply(
    lambda row: "Perf_shallower" if row['top_prf_int'] >= (row['Depth_2top_Eclay'] + row['Thickness_Eclay']) else "Perf_deeper",
    axis=1
)

merged_df.head()
```

```python
# Convert `msmt_date` to datetime format with error handling for inconsistent formats
merged_df['msmt_date'] = pd.to_datetime(merged_df['msmt_date'], errors='coerce', infer_datetime_format=True)

# Ensure numeric conversion for relevant columns
for col in ['altitude_top_prf', 'altitude_top_EClay', 'altitude_base_EClay', 'wlm_gse', 'altitude_bot_prf', 'gse_gwe',
           'altitude_top_Cclay','Cclay_base_altitude','altitude_top_Aclay','Aclay_base_altitude',
            'altitude_top_MidTulareClay','altitude_base_MidTulareClay','altitude_top_AmnicolaClaystone',
            'altitude_base_AmnicolaClaystone',
           ]:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

# Drop rows with NaT in `msmt_date` (optional, if you want to ignore rows with invalid dates)
merged_df = merged_df.dropna(subset=['msmt_date'])

# Create a PDF file to save the plots
with PdfPages('GoundwaterElevations_Perforations_ConfiningUnits_plots.pdf') as pdf:
    # Group data by `site_code` and create a time series plot for each group
    for site, group_data in merged_df.groupby('site_code'):
        plt.figure(figsize=(10, 6))

        # Main groundwater elevation plot
        plt.plot(group_data['msmt_date'], group_data['gse_gwe'], marker='o', linestyle='-', label=f'Site: {site} (gse_gwe)')

        # Dashed line for `wlm_gse` in blue with linewidth 1.5
        plt.plot(group_data['msmt_date'], group_data['wlm_gse'], linestyle='--', color='blue', linewidth=1.5, label='wlm_gse')

        # Dashed line for `top_prf_int` in green with linewidth 1
        plt.plot(group_data['msmt_date'], group_data['altitude_top_prf'], linestyle='--', color='green', linewidth=3, label='altitude_top_prf')

        # Dashed line for `bot_prf_int` in red with linewidth 0.8
        plt.plot(group_data['msmt_date'], group_data['altitude_bot_prf'], linestyle='--', color='red', linewidth=3 , label='altitude_bot_prf')

        # Solid line with square markers for `Depth to the Top of Corcoran`
        plt.plot(group_data['msmt_date'], group_data['altitude_top_EClay'], linestyle='--', marker='s', color='brown', linewidth=2, label='altitude_top_Eclay')

        # Solid line with triangle markers for `Depth to the Base of Corcoran`
        plt.plot(group_data['msmt_date'], group_data['altitude_base_EClay'], linestyle='--', marker='s', color='brown', linewidth=2, label='altitude_base_Eclay')        
       
        # Solid line with square markers for `Depth to the Top of Cclay`
        plt.plot(group_data['msmt_date'], group_data['altitude_top_Cclay'], linestyle='--', marker='*', color='green', linewidth=2, label='altitude_top_Cclay')

        # Solid line with square markers for `Depth to the Base of Cclay`
        plt.plot(group_data['msmt_date'], group_data['Cclay_base_altitude'], linestyle='--', marker='*', color='green', linewidth=2, label='Cclay_base_altitude')

        # Solid line with square markers for `Depth to the Top of Aclay`
        plt.plot(group_data['msmt_date'], group_data['altitude_top_Aclay'], linestyle='--', marker='o', color='red', linewidth=2, label='altitude_top_Aclay')

        # Solid line with square markers for `Depth to the Base of Aclay`
        plt.plot(group_data['msmt_date'], group_data['Aclay_base_altitude'], linestyle='--', marker='o', color='red', linewidth=2, label='Aclay_base_altitude')

        # Solid line with square markers for `Depth to the Top of MidTulare claystone`
        plt.plot(group_data['msmt_date'], group_data['altitude_top_MidTulareClay'], linestyle='--', marker='o', color='black', linewidth=2, label='MidTulareClay_base_altitude')
        
        # Solid line with square markers for `Depth to the Base of MidTulare claystone`
        plt.plot(group_data['msmt_date'], group_data['altitude_base_MidTulareClay'], linestyle='--', marker='o', color='black', linewidth=2, label='MidTulareClay_base_altitude')

        # Solid line with square markers for `Depth to the Top of Amnicola claystone`
        plt.plot(group_data['msmt_date'], group_data['altitude_top_AmnicolaClaystone'], linestyle='--', marker='o', color='blue', linewidth=2, label='AmnicolaClay_base_altitude')
        
        # Solid line with square markers for `Depth to the Base of Amnicola claystone`
        plt.plot(group_data['msmt_date'], group_data['altitude_base_AmnicolaClaystone'], linestyle='--', marker='o', color='blue', linewidth=2, label='AmnicolaClay_base_altitude')

        # Add title with `Eclay` as part of the title
        eclay_label = group_data['Eclay'].iloc[0] if 'Eclay' in group_data.columns else 'N/A'
        plt.title(f'Groundwater Elevation Time Series for Site {site} - Eclay: {eclay_label}')

        # Set labels and other plot settings
        plt.xlabel('Date')
        plt.ylabel('Groundwater Elevation (gse_gwe)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()  # Ensures the layout fits well
               
        # Save the current figure to the PDF
        pdf.savefig()  # This saves the current figure to the PDF
        plt.close()    # Close the figure to free memory

print("Plots have been saved to 'groundwater_elevation_plots.pdf'")
```
